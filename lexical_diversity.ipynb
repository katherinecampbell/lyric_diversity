### Firstly: Current

#### Number 1: Tesselate by Alt-J

```python
#function to find lyrical diversity
def lexical_diversity(text):
    print (len(set(text)))
    print (len(text))
    return len(set(text)) / len(text)
```

```python
#function to import and read from txt file
def load_txt(file):
    with open(file, encoding="utf8") as f:
        return ' '.join(line.strip() for line in f.readlines())
```

```python
#define punctuation as to preserve words with apostrophes - very common in lyrics
punc = str.maketrans('', '', string.punctuation)
```

```python
#access URL and use beautifulsoup parser
MAX_RETRIES = 20
url = 'http://www.songtexte.com/songtext/alt-j/tessellate-3b0f5c7.html'
```

```python
session = requests.Session()
adapter = requests.adapters.HTTPAdapter(max_retries=MAX_RETRIES)
session.mount('https://', adapter)
session.mount('http://', adapter)
page = session.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
lyrics = soup.find_all('div', id='lyrics')
```

```python
#make result set a list
altj = [div.get_text() for div in lyrics]
```

```python
#function to clean and normalize all lyrics from beautiful soup parser
def clean_soup(lyrics, string):
    #make list string
    lyrics = str(lyrics)
    lyrics= lyrics.strip('[]')
    #erase new lines
    lyrics = lyrics.replace('\\n', " ")
    lyrics = lyrics.replace("\\", "" )
    lyrics = lyrics.replace('?', ' ')
    #erase punctuation
    words = lyrics.split()
    cleaning = [ word.translate(string) for word in words]
    #make all lowercase to normalize
    cleaning = [word.lower() for word in cleaning]
    txt = ' '.join(cleaning)
    txt = nltk.word_tokenize(txt)
    print(txt)
    return lexical_diversity(txt)
```

```python
clean_soup(altj, punc)
```

### The Classics: The Who and the Rolling Stones - beautifulsoup parser

#### Number 3: Who are You? by The Who

```python
MAX_RETRIES = 20
url = 'http://www.lyricsfreak.com/w/who/who+are+you_20146820.html'
```

```python
session = requests.Session()
adapter = requests.adapters.HTTPAdapter(max_retries=MAX_RETRIES)
session.mount('https://', adapter)
session.mount('http://', adapter)
page = session.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
lyrics = soup.find_all('div',id='content_h')
```

```python
who = [div.get_text() for div in lyrics]
```

```python
#fix anomalies from scraping
new_list = []
repls = ('x11Who', 'who'),('PinI','pin i'),('youWho', 'you who'), ('holeYet','hole yet'),('kissHow','kiss how'), ('treesMy', 'trees my'),('dayEleven', 'day eleven'),('clownWith','clown with'),('hiccuppedAnd','hiccuped and'),('undergroundAnd', 'underground and'),('townBack', 'town back'), ('walkedWhere', 'walked where'), ('TinI', 'tin i'), ('doorwayA', 'doorway a'), ('nameHe', 'name he'), ('hairI', 'hair i'), ('aroundAnd', 'around and'), ('wayWell', 'way well'),('PanGod', 'pan god'),('knowC\'mon', 'know cmon'), ('nowAfter', 'now after'),('walkedWhere', 'walked where'), ('yeahYou', 'yeah you'),('knowOh', 'know oh'), ('chairWell', 'chair well'), ('knowsYou', 'knows you')
for item in who:
    item = reduce(lambda a, kv: a.replace(*kv), repls, item)
    new_list.append(item)
```

```python
clean_soup(new_list, punc)
```
